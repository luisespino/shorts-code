```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

X = np.array([
    [2, 4, 6],
    [3, 6, 5],
    [4, 3, 8],
    [5, 7, 6],
    [6, 5, 7],
    [7, 8, 5],
    [8, 6, 9],
    [9, 5, 6]
])

fig = plt.figure(figsize=(3.4, 5))
ax = fig.add_subplot(projection='3d')
ax.set_title("Datos originales 3D")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], color='blue', label='Originales')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python


pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

fig = plt.figure(figsize=(3, 3))
ax = fig.add_subplot()
ax.set_title("Datos con PCA 2D")
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], color='green', label='Transformados')
ax.set_xlabel('Feature1')
ax.set_ylabel('Feature2')
ax.axhline(0, color='gray', linestyle='--', linewidth=0.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
