```python

import numpy as np
from sklearn.metrics \
  import accuracy_score
from sklearn.naive_bayes \
  import GaussianNB

np.random.seed(42)
s = 100
f = 4

# Features aleatorias
X = np.random.randn(s,f)

# Etiquetas aleatorias
y = np.random.randint(0, 2, size=s)

# entrenamiento
nb = GaussianNB()
nb.fit(X, y)

# Predicci贸n y evaluaci贸n
y_pred = nb.predict(X)

accuracy = accuracy_score(y, \
                          y_pred)
print(f"Precisi贸n del modelo: \
          {accuracy * 100:.2f}%")
```
