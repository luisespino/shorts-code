Import

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a)
```

Slicing

```python
print(a[1:3])
print(a[3:])
print(a.reshape(-1,1))
```

Métodos aritméticos

```python
print(a.sum())
print(a.min())
print(a.max())
```

Métodos estadísticos

```python
print(a.mean())
print(a.std())
print(a.var())
```

Funciones

```python
print(np.average(a, weights=a))
print(np.percentile(a,75))
```
