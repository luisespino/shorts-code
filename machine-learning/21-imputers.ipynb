```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Edad': [
        25,
        30,
        np.nan,
        45,
        40],
    'Ingresos': [
        50000,
        np.nan,
        60000,
        80000,
        np.nan]
})

print("Datos originales:")
print(data)
```

```python
from sklearn.impute import \
    SimpleImputer

imputer = \
    SimpleImputer(strategy='mean')
data_imputada = \
    imputer.fit_transform(data)
data_imputada = \
    pd.DataFrame(data_imputada, \
                 columns=data.columns)

print("\nDatos imputados:")
print(data_imputada)
```
