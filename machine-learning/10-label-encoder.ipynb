```python

from sklearn import preprocessing

a = ['false', 'false', 'true']
b = ['normal', 'high', 'high']
c = ['cool', 'hot', 'cool']
d = ['aaaa', 'bbbb', 'cccc']

out = ['yes', 'yes', 'no']

enc = preprocessing.LabelEncoder()

ae = enc.fit_transform(a)
be = enc.fit_transform(b)
ce = enc.fit_transform(c)
de = enc.fit_transform(d)
oute = enc.fit_transform(out)

print(ae)
print(be)
print(ce)
print(oute)

features = list(zip(ae.tolist(), \
                    be.tolist(), \
                    ce.tolist(), \
                    de.tolist()))
print(features)
```
