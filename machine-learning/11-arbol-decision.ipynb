```python


import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.tree import \
  DecisionTreeClassifier, \
  plot_tree

# dataset de Quinlan
outlook=['sunny','sunny','overcast','rain','rain','rain','overcast','sunny','sunny','rain','sunny','overcast','overcast','rain']
tempe=['hot','hot','hot','mild','cool','cool','cool','mild','cool','mild','mild','mild','hot','mild']
humi=['high','high','high','high','normal','normal','normal','high','normal','normal','normal','high','normal','high']
windy=['false','true','false','false','false','true','true','false','false','false','true','true','false','true']

play=['N','N','P','P','P','N','P','N','P','P','P','P','P','N']

# LabelEncoder
e = preprocessing.LabelEncoder()
outlook_e=e.fit_transform(outlook)
tempe_e=e.fit_transform(tempe)
humidity_e=e.fit_transform(humi)
windy_e=e.fit_transform(windy)
label=e.fit_transform(play)

# lista de tuplas
features=list(zip(outlook_e,\
                  tempe_e,\
                  humidity_e,\
                  windy_e))

# árbol de decisión
model = DecisionTreeClassifier()
model.fit(features, label)
plt.figure(figsize=(3.5, 7))
plot_tree(model, filled=True, \
          class_names =['N','P'],\
          label='root', fontsize=6)
plt.show()
```
