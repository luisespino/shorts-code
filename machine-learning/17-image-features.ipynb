```python


from PIL import Image, ImageDraw, \
    ImageFont
import matplotlib.pyplot as plt
import numpy as np

# Create image
image = Image.new('L', (20, 20), \
                  color=0)
draw = ImageDraw.Draw(image)
draw.text((7, 4), "4", fill=255)

# Show original image
print("Original image:")
display(image)

# Extract features from image
array_image = np.array(image)
f = np.unique(array_image)
print(f"Features: {f}")

# show image with plt
plt.figure(figsize=(3, 3))
plt.imshow(array_image, \
           cmap='gray')
plt.colorbar()
plt.show()
```
